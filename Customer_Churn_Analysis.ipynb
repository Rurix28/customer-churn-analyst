```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Churn'].value_counts(normalize=True)
```

```python
df.groupby('Contract')['Churn'].value_counts(normalize=True).unstack().plot(kind='bar')
plt.title("Churn Rate by Contract")
plt.show()
```

```python
sns.boxplot(x='Churn', y='tenure', data=df)
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.show()
```

Insigth pada Customer Chrun Analyst tersebut menunjukkan bahwa Pelanggan dengan kontrak month-to-month memiliki churn rate sekitar 4 kali lebih tinggi dibandingkan dengan pelanggan kontrak 1 tahun ataupun lebih.
